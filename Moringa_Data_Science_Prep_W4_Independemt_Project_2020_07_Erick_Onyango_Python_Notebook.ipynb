## 1.4 Cleaning our Dataset

### 1.41) Validity 

```
#Dropping columns with missing values
df= df.drop(['Scheduled at','Displayed comment'], axis=1)
df
```

```
# Procedure 2: Column that are similar
# Data Cleaning Action: Drop cars attribute
# Explanation: Its irrelevant since it is simillar to bluercar counter
# 
df =df.drop(['Cars'],axis =1)
df
```

### 1.42) Accuracy 

```
# Procedure 1: Creating the datetime column by combining the date and time
# Data Cleaning Action: Concatenating the four paramenters
# Explanation: for ease of analysis
#
df['Datatime'] =pd.to_datetime(df[['year','month','day','hour','minute']])
```

### 1.43) Completeness

```
# Procedure 1:removing unncessary columns
# Data Cleaning Action: dropping the Charging status, rental status and subscription status
# Explanation: there are pointing to the same ting and without any much meaning
df = df.drop(['Charging Status','Geo point','Rental status','Subscription status'],axis = 1)
df
```

### 1.44) Consitency: 

```
#Renaming the columns to remove the ones with spaces in between
#consistency in naming
df.columns = ['Address','Bluecar','Utilib','Utilib14','Slots','City','Id','Kind','Postalcode','PublicName','Slots2','StationType','Status','Year','Month','Day','Hour','Minute','Datetime']
df
```

* FINAL DATA

```
# Getting the data belonging to Paris city
parisdata = df[df.City =='Paris']
parisdata
```

```
# sorting Up the paris 
parisdata = parisdata.sort_values(by=['Datetime'],ascending = True)
parisdata
```

## 1.5 Answering Questions

Let's answer the following questions with our dataset using python.

```
# What is the most popular hour of picking up the blue Car

# getting the differnce of blue cars
# any negative deviation is an indicator that the car is picked
parisdata['Deviation'] = parisdata.Bluecar.diff()
```

```
# Getting the hour difference
parisdata10 = parisdata[parisdata.Deviation <  0].groupby(['Hour'])['Deviation'].sum()
parisdata10.sort_values(ascending =False).head(1)
#
# 6:00pm is when most of the cars are picked from the station in paris

```

```
# What is the time in which most of the cars were returned
#
parisdata10 = parisdata[parisdata.Deviation >  0].groupby(['Hour'])['Deviation'].sum()
parisdata10.sort_values(ascending =False).head(1)
#
# 8:00am is the time in which most car were returned
```

```
# What is the most popular station -Overal
parisdata2 = parisdata['Id'].value_counts().head(1)
parisdata2
#paris-portedenttrouge-8 is the most popular station
```

```
# What is the most populaR postal code for picking up the Bluecar
parisdata3 = parisdata[(parisdata.Deviation < 0)].groupby(['Postalcode'])['Deviation'].sum()
parisdata3.sort_values(ascending = False).head(1)

```
